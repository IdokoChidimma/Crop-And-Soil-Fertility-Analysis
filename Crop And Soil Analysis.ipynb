```python
import numpy as np
```

```python
np.random.seed(0)
data["pH"] = np.random.uniform(5.5, 8.0, size=len(data))
```

```python
def classify_pH(pH):
    if pH < 6.5:
        return "Acidic"
    elif pH <= 7.5:
        return "Neutral"
    else:
        return "Alkaline"
data["pH Type"] = data["pH"].apply(classify_pH)
```

```python
#WHICH CROPS PREFER NEUTRAL SOIL?
```

```python
data[data["pH Type"] =="Neutral"]["Crop Type"].value_counts()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
os.makedirs("images", exist_ok=True)
```

```python
data.groupby("Crop Type")[["Nitrogen", "Phosphorous", "Potassium"]].mean().plot(kind='bar', figsize=(14, 6))
plt.title("Average NPK Levels per Crop")
plt.ylabel("Level")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("images/Average NPK Levels per Crop.png", )
plt.show()
```

```python
sns.countplot(data=data, x='Crop Type', order=data['Crop Type'].value_counts().index)
plt.xticks(rotation=45)
plt.title('Distribution of Crops in Dataset')
plt.savefig('images/crop_distribution.png')
plt.show()
```

```python

sns.countplot(data=data, x='Soil Type', order=data['Soil Type'].value_counts().index)
plt.xticks(rotation=45)
plt.title('Distribution of Soils in Dataset')
plt.show()
```

```python
sns.histplot(data['pH'], bins=20, kde=True, color='green')
plt.title('Distribution of Soil pH')
plt.xlabel('Soil pH')
plt.ylabel('frequency')
plt.savefig('images/Distribution of soil.png')
plt.show()
```

```python
environmental_cols = ["Temparature", "Humidity","Moisture"]
nutrient_cols = ["Nitrogen", "Phosphorous", "Potassium"]
all_relevant_cols = environmental_cols + nutrient_cols
```

```python
correlation_matrix = data[all_relevant_cols].corr()
```

```python
plt.figure(figsize=(14, 7))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm",fmt=".2f", linewidths=.5)
plt.title("Correlation Between Environmental and Nutrient Factors")
plt.savefig('images/correlation_heatmap.png')
plt.show()
```

```python
plt.figure(figsize=(12,7))
sns.barplot(x="Crop Type", y="Nitrogen", data=data)
plt.title("NItrogen Levels Across Different Crop Types")
plt.xticks(rotation=45)
plt.savefig('images/nitrogen_boxplot.png')
plt.show()
```

```python


plt.figure(figsize=(14,7))
sns.barplot(x="Crop Type", y="Phosphorous", data=data)
plt.title("Phosphorous Levels Across Different Crop Types")
plt.xticks(rotation=90)
plt.savefig('images/Phosphorous_boxplot.png')
plt.show()
```
